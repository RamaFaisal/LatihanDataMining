```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
data = pd.read_csv('Invistico_Airline.csv')

# Menampilkan statistik deskriptif
print(data.describe())
```

```python
# Visualisasi distribusi usia dan jarak penerbangan
sns.histplot(data['Age'], kde=True)
plt.show()
sns.boxplot(x='Satisfaction', y='Flight_Distance', data=data)
plt.show()
```
